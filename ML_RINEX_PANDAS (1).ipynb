```python
df_sample = pd.read_csv('nba_sample.csv')
df_sample
```

```python
#Calculate the BMI of the players
#BMI=Weight(kg)/Height(m^2)
df_sample["player_height_m"] = df_sample['player_height']/100
df_sample["BMI"] = df_sample['player_weight']/(df_sample['player_height_m']**2)
```

```python
'''Create a new column with the name BMI Evaluation. Assign the BMI status based on the following condition,


BMI <= 18.4 - "Underweight"

BMI <= 24.9 - "Healthy"

BMI <= 29.9 - "Over Weight"

BMI <= 34.9 - "Severely Over Weight"

BMI <= 39.9 - "Obese"
39.9 - "Severely Obese."
'''

def evaluate_bmi(bmi):
    if bmi <= 18.4:
        return "Underweight"
    elif bmi <= 24.9:
        return "Healthy"
    elif bmi <= 29.9:
        return "Overweight"
    elif bmi <= 34.9:
        return "Severely Overweight"
    elif bmi <= 39.9:
        return "Obese"
    else:
        return "Severely Obese"

# Apply the function to create the "BMI Evaluation" column
df_sample["BMI Evaluation"] = df_sample["BMI"].apply(evaluate_bmi)
df_sample
```

```python
#Group by function: Can be used in association with the aggregate functions
df.groupby(by=['Pclass',"Sex"]).count()
```

```python
#Sum of fare spent by females according to the specific classes
df.groupby(by=['Pclass','Sex'])['Fare'].sum()
```

```python
#Mean of fare spent by females according to the specific classes sorted in ascending order
df.groupby(by=['Pclass','Sex'])['Fare'].mean().sort_values()
```

```python
#To specify multiple aggregation functions
df.groupby(by=['Pclass','Sex'])['Fare'].agg(['max','min','sum','mean','count'])
```

```python
#To add indices
df.groupby(by=['Pclass','Sex'])['Fare'].agg(['max','min','sum','mean','count']).reset_index()
```

```python
#Specifying the aggregation functions for the columns
df.groupby(by=["Pclass",'Sex']).agg({"Age":"mean","Fare":"sum","PassengerId":"count"})
```
