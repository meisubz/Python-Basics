```python
#Data Visualization using Matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Line Plot
x = [1,2,3,4,5]
y = [10,9,8,7,6]
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Random Dancing Numbers',c='blue',fontsize=18)
plt.show()
```

```python
#Line Plot with Numpy Array
x = np.array([55,66,77,88,99])
y = x*2
plt.plot(x,y)
plt.xlabel('Large Numbers')
plt.ylabel('Twice the Large Numbers')
plt.title('Just Numbers',c='red')
plt.show()
```

```python
#Multiple Plots on the Same Axis: Basically adding two plots within the same axis
x = np.array([11,12,13,14,15])
y = x*10
plt.plot(x,y,"-*")
x1 = np.array([5,6,7,8,9])
y1 = x1*10
plt.plot(x1,y1,"-*")
plt.xlabel('X-Array')
plt.ylabel('Y-Array')
plt.title('Line Plot With Multiple Plots On The Same Axis')
plt.show()
```

```python
#Filling the area between two plots
#Multiple Plots on the Same Axis: Basically adding two plots within the same axis
x = np.array([2,4,6,8])
y = x*2
plt.plot(x,y,"--",color='red')
x1 = np.array([2,4,6,8])
y1 = x1*5
plt.plot(x1,y1,"-*",color='blue')
plt.xlabel('X-Array')
plt.ylabel('Y-Array')
plt.title('Line Plot With Multiple Plots On The Same Axis')
plt.fill_between(x,y,y1,color='green',alpha=0.3)
plt.show()
```

```python
x = np.array([11,12,13,14,15])
y = x*10
plt.plot(x,y,"-*",c='pink')
x1 = np.array([5,6,7,8,9,10])
y1 = x1*10
plt.plot(x1,y1,"-*",c='purple')
x2 = np.array([0,1,2,3,4])
y2 = x2*10
plt.plot(x2,y2,'-+',c='yellow')
plt.xlabel('X-Array')
plt.ylabel('Y-Array')
plt.title('Line Plot With Multiple Plots On The Same Axis')
plt.show()
```

```python
#Scatter Plot
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.scatter(x,y)
```

```python
#Scatter Plot with different shape & color for two datasets

#DS-1
x1 = [89,43,36,95,10,66,34,38,20,6]
y1 = [21,46,3,35,67,95,53,72,58,10]

#DS-2
x2 = [26,29,48,64,6,5,36,66,72,10]
y2 = [26,24,90,33,38,20,56,2,47,15]

plt.scatter(x1,y1,c='yellow',linewidth=3,edgecolors='red',s=50,marker='D')
plt.scatter(x2,y2,c='pink',linewidth=3,edgecolors='blue',s=50,marker='h')
plt.show()
```

```python
#Bar Chart
x = ['A','B','C','D']
y = [5,6,7,8]
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Bar Graph')
plt.bar(x,y,width=0.5)
plt.show()
```

```python
#Horizontal Bar Chart
x = ['A','B','C','D']
y = [5,6,7,8]
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Bar Graph')
plt.barh(x,y,height=0.5,color='pink',edgecolor='purple')
plt.show()
```

```python
#Pie Chart
categories = ['Apple','Banana','Guava','Peach','Plum'] #Taken as labels
data = [10,20,30,40,50]
exp=[0,0,0,0.2,0] #Parameters to "explode" the
plt.pie(data,labels = categories, autopct='%1.1f%%',explode=exp) #autopct is used to display the data in terms of percentage, colors can be used in a list to specify the data in the set
plt.show()
```
